```python
repo_title=[]
repo_desc=[]
cont_cnt=[]
lang=[]
```

```python
try:
    rt = driver.find_elements(By.CLASS_NAME,'v-align-middle')
    for i in rt:
        repo_title.append(i.text)
        
    rd = driver.find_elements(By.CLASS_NAME,'mb-1')
    for i in rd:
        repo_desc.append(i.text)
        
    cc = driver.find_elements(By.XPATH,'//a[@class="Link--muted"][1]')
    for i in cc:
        cont_cnt.append(i.text)
        
    ln = driver.find_elements(By.XPATH,'//span[@itemprop="programmingLanguage"][1]')
    for i in ln:
        lang.append(i.text)
        
except NoSuchElementException:
    pass
```

```python
for i in range(len(repo_title)):
    if repo_title[i]=='':
        repo_title[i]='-'
    else:
        repo_title.append('-')
    if repo_desc[i]=='':
        repo_desc[i]='-'
    else:
        repo_desc.append('-')
    if cont_cnt[i]=='':
        cont_cnt[i]='-'
    else:
        cont_cnt.append('-')
    if lang[i]=='':
        lang[i]='-'
    else:
        lang.append('-')
lang.append('-')
lang.append('-')
lang.append('-')
```

```python
print(len(repo_title),len(repo_desc),len(cont_cnt),len(lang))
```

```python
import pandas as pd
df=pd.DataFrame({'Repository Title':repo_title,'Repository Description':repo_desc,'Contributors Count':cont_cnt,'Langauge':lang})
df
```
