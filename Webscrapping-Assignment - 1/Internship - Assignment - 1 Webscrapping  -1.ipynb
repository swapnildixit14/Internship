Q2 . Assignment 1 - IMDB 50 Movies

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page2=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
page2
```

```python
soup2=BeautifulSoup(page2.content)
```

```python
soup2
```

```python
mname2 = [] #empty movies title list
for i in soup2.find_all('td',class_="titleColumn"):
    mname2.append(i.text)
mname2[0:51]
```

```python
rating2 = [] #empty list for ratings
for i in soup2.find_all('td',class_="ratingColumn imdbRating"):
    rating2.append(i.text)
rating2[0:51]
    
```

```python
rel_year2 = [] #empty list for release year
for i in soup2.find_all('span',class_="secondaryInfo"):
    rel_year2.append(i.text)
rel_year2[0:51]
```

```python
print(len(mname2),len(rating2),len(rel_year2))
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':mname2,'Release Year':rel_year2,'Ratings':rating2})
df
```

Q7. Assignment - 1 News Details

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page7=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page7
```

```python
soup7=BeautifulSoup(page7.content)
soup7
```

```python
title7 = [] #Empty list of news
for i in soup7.find_all('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"):
    title7.append(i.text)
title7
    
```

```python
time7 = [] #Empty list of news
for i in soup7.find_all('span',class_="RiverByline-datePublished"):
    time7.append(i.text)
time7
    
```

```python
link7 = [] #Empty list of news
for i in soup7.find_all('span',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"): 
    link7.append(i.get('href'))
link7
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':title7,'Time':time7,'Links':link7})
df
```

Q8 - Assignment - 1 - Author

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page8=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page8
```

```python
soup8=BeautifulSoup(page8.content)
soup8
```

```python
#Paper Title
#Authors
#Published Date
#Paper URL

paper_title = [] #empty list for paper title
for i in soup8.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    paper_title.append(i.text)
paper_title
```

```python
author = [] #empty list for paper author
for i in soup8.find_all('span',class_='sc-1w3fpd7-0 dnCnAO'):
    author.append(i.text)
author
```

```python
published_date = [] #empty list for published date
for i in soup8.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    published_date.append(i.text)
published_date
```

```python
paper_url = [] #empty list for paper url
for i in soup8.find_all('a',class_='sc-5smygv-0 fIXTHm'):
    paper_url.append(i.get('href'))
paper_url
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':paper_title,'Author':author,'Published Date':published_date,'URL':'paper_url'})
df
```
