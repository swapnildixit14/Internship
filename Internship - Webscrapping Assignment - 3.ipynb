“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver = webdriver.Chrome(r"D:\Swapnil Dixit\Data Science\Data Trained\Internship -Fliprobo\Tasks\03 April 2023\chromedriver_win32\chromedriver.exe")
driver.get('https://www.flipkart.com/')
```

```python
#finding the element for product search bar
designation = driver.find_element(By.CLASS_NAME,"_3704LK")
designation.send_keys('smartphone')
```

```python
#Search button click
search_btn=driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]')
search_btn.click()
```

```python
brand_name=[]
smartphone_name=[]
colour=[]
ram=[]
rom=[]
pcamera=[]
scamera=[]
display_size=[]
battery=[]
price=[]
purl=[]
```

```python
#Scrapping for brand
null= " "
brand_tags=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in brand_tags[0:10]:
    bt=i.text
    if (bt==" "):
        bt="-"
    brand_name.append(bt)
    
#Scrapping product smartphone name
smart_tags=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in smart_tags[0:10]:
    st=i.text
    if (st==" "):
        st="-"
    smartphone_name.append(st)
    
#Scrapping product smartphone colour
color_tags=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in color_tags[0:10]:
    ct=i.text
    if (ct==null):
        st="-"
    colour.append(ct)
    
#Scrapping product smartphone ram
ram_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in ram_tags[0:10]:
    rt=i.text
    if (rt==null):
        rt="-"
    ram.append(rt)
    
#Scrapping product smartphone ram
rom_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in rom_tags[0:10]:
    rot=i.text
    if (rot==null):
        rot="-"
    rom.append(rot)

#Scrapping product smartphone primary camera
pcamera_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in pcamera_tags[0:10]:
    pc=i.text
    if (pc==null):
        pc="-"
    pcamera.append(pc)
    
#Scrapping product smartphone Secondary camera
scamera_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in scamera_tags[0:10]:
    sc=i.text
    if (sc==null):
        sc="-"
    scamera.append(sc)
    
#Scrapping product smartphone Display Size
dsize_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in dsize_tags[0:10]:
    ds=i.text
    if (ds==null):
        ds="-"
    display_size.append(ds)
    
#Scrapping product smartphone Battery
battery_tags=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"]')
for i in battery_tags[0:10]:
    bt=i.text
    if (bt==null):
        bt="-"
    battery.append(bt)
    
#Scrapping product smartphone Price
price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in price_tags[0:10]:
    pr=i.text
    if (pr==null):
        pr="-"
    price.append(pr)
    
#Scrapping product smartphone Product URL
purl_tags=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in purl_tags[0:10]:
    if i.text==null:
        i.text="-"
    purl.append(i.get_attribute("href"))
```

Q6. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver = webdriver.Chrome(r"D:\Swapnil Dixit\Data Science\Data Trained\Internship -Fliprobo\Tasks\03 April 2023\chromedriver_win32\chromedriver.exe")
driver.get('https://www.digit.in/laptops/asus-tuf-gaming-fx505-price-173693.html')
```

```python
brand_name=[]
market_status=[]
release_date=[]
digit_ratings=[]
user_ratings=[]
os=[]
price=[]
```

```python
#Scrapping for brand
brand_tags=driver.find_elements(By.XPATH,'//div[@class="heading-wraper"]')
for i in brand_tags:
    bt=i.text
    brand_name.append(bt)
    
#Market Status

market=driver.find_elements(By.XPATH,'//strong[@class="red"]')
for i in market:
    m=i.text
    release_date.append(m)
    
#Release_date
rd=driver.find_elements(By.XPATH,'//strong[@class="red"]')
for i in rd:
    bt=i.text
    release_date.append(rd)   
    
#Digit Rating
dr=driver.find_elements(By.XPATH,'//span[@class="rating_no yellowtext"]')
for i in dr:
    bt=i.text
    digit_ratings.append(dr)   
    
    
#User Rating
ur=driver.find_elements(By.XPATH,'//span[@class="rating_no greentext"]')
for i in ur:
    bt=i.text
    user_ratings.append(ur) 
    
    
#Operating System
ops=driver.find_elements(By.XPATH,'//div[@class="value"]')
for i in ops:
    ot=i.text
    os.append(ot)

#Price
p=driver.find_elements(By.XPATH,'//div[@class="price"]')
for i in p:
    pr=i.text
    price.append(pr)
    
print("\nBarnd Name",brand_name)
print("\nMarket Status",market_status)
print("\nRelease Date",release_date)
print("\nDigit Ratings",digit_ratings)
print("\nUser Ratings",user_ratings)
print("\nOperating System",os)
print("\Price",price)
```

Q9 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver = webdriver.Chrome(r"D:\Swapnil Dixit\Data Science\Data Trained\Internship -Fliprobo\Tasks\03 April 2023\chromedriver_win32\chromedriver.exe")
driver.get('https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2023-04-20&to=2023-04-23&guests=2&page=1')
```

```python
hostel_name=[]
dist=[]
ratings=[]
total_review=[]
pprice=[]
dprice=[]
```

```python
#Scrapping for Hostel Name
v1=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
for i in v1[0:10]:
    bt1=i.text
    hostel_name.append(bt1)

#Scrapping for Distance
v1=driver.find_elements(By.XPATH,'//span[@class="description"]')
for i in v1[0:10]:
    bt1=i.text
    dist.append(bt1)
    
#Scrapping for Ratings
v1=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
for i in v1[0:10]:
    bt1=i.text
    ratings.append(bt1)
    
    
#Scrapping for Total Reviews
v1=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
for i in v1[0:10]:
    bt1=i.text
    total_review.append(bt1)
    
#Scrapping for Total Private Price
v1=driver.find_elements(By.XPATH,'//p[@class="no-price body-3 no-privates"]')
for i in v1[0:10]:
    bt1=i.text
    pprice.append(bt1)

#Scrapping for Total Dorms Price
v1=driver.find_elements(By.XPATH,'//div[@class="price title-5"]')
for i in v1[0:10]:
    bt1=i.text
    dprice.append(bt1)
    
print("\n Hostel Name = ", hostel_name)
print("\n Distance = ", dist)
print("\n Ratings = ", ratings)
print("\n Total Review = ", total_review)
print("\n Private Price = ", pprice)
print("\n Dorms Price = ", dprice)
```
