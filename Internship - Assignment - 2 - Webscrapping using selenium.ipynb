Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
url = "https://www.naukri.com"
driver.get(url)
```

```python
#finding the element for job search bar
designation = driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scrapping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
#Scrapping job location from the given page
location_tag=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tag[0:10]:
    loc=i.text
    job_location.append(loc)

#Scrapping company name from the given page
company_tags=driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
    
#Scrapping job experience from the given page
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
                                                                
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import pandas as pd
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company_name':company_name,'Experience':experience_required})
df
```

3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
url="https://www.azquotes.com/top_quotes.html"
driver.get(url)
```

```python
Quote=[]
Author=[]
Type=[]
```

```python
#Scrapping for Quote
quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote_tags[0:1000]:
    qt=i.text
    Quote.append(qt)
    
#Scrapping Author
author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in author_tags[0:1000]:
    at=i.text
    Author.append(at)


#Type of quotes
type_quotes=driver.find_elements(By.XPATH,"//div[@class='tags']")
for i in type_quotes[0:1000]:
    tq=i.text
    Type.append(tq)
```

```python
print(len(Quote),len(Author),len(Type))
```

```python
import pandas as pd
df=pd.DataFrame({'Quote':Quote,'Author':Author,'Type':Type})
df
```

Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpage https://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make the DataFrame.

```python
url="https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"
driver.get(url)
```

```python
pname=[]
born_dead=[]
term=[]
remark=[]
```

```python
#Scrapping for Pname
Pn=driver.find_elements(By.XPATH,'//p[@class=""]')
for i in Pn:
    qt=i.text
    pname.append(qt)
    
#Scrapping born dead
bd_tags=driver.find_elements(By.XPATH,'//div[@class=""]')
for i in bd_tags:
    at=i.text
    born_dead.append(at)


#Type of term
term_tags=driver.find_elements(By.XPATH,"//span[@class='']")
for i in term_tags:
    tq=i.text
    term.append(tq)

    #Type of remark
remark_tags=driver.find_elements(By.XPATH,"//p[@class='']")
for i in remark_tags:
    tq=i.text
    remark.append(tq)
```

```python
print(len(pname),len(born_dead),len(term),len(remark))
```

```python
import pandas as pd
df=pd.DataFrame({'Prime Ministers Name':pname,'Born Dead':born_dead,'Term':term,'Remark':remark})
df
```
